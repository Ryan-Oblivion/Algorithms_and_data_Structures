```python
# problem 2

my_file = open('large.graph', 'r')

#this helps display the file
content_file = my_file.readlines()

# now i realized there were unwanted tabs and new lines in the file so i joined the list into a string to start removing
joinedg =''.join(content_file)

#this replaced the tab with a colon
s = joinedg.replace('\t', ':')

#this replaced the new line with a comma
m = s.replace('\n',',')

# this made my string back into a list and split at the comma
l = m.split(',')
# needed to remove the last element because it was empty and caused problems creating a dict later
l.pop(-1)

#i wanted to then make my list of strings into a list of tuples. this helps to understand the data better
# it may also make it easier to run a bfs easier also
# this then further used the colon to split into a tuple
lt =[tuple(map(str, sub.split(':'))) for sub in l]
# next i wanted to turn everything into a dict
lt_d = {}


for (key, value) in lt:
    lt_d.setdefault(key,[]).append(value)

import math    # i needed to get inf for the distance so i imported math

colors = {}       # then outside of the bfs function i created empty dictionaries of colors, d, and parents to populate them later
d = {}
parents = {}
def BFS(lt_d, u):       # here is the begining of the bfs function, using the adjacency list created from the graph
    for u in lt_d:
        colors[u] = 'white';     # next to initialize the verticies where they start white, with no parents at position u
        d[u] = math.inf;
        parents[u] = None;
    
    colors[u] = 'gray'; d[u] = 0; parents[u] == None;  # here they will turn the neighbors gray and prepare to append u to the queue, which is enqueue
    Q = []; Q.append(u);
    
    while Q: #!= {}: i commented this part and it avoids the error  # while the queue is not empty we will remove the first element 
        u = Q.pop(0); # i see that the code works, but once the queue is empty we get an error. this dequeues from Q
        print(u, end=' ')  # i added this code to see the bfs in action. just comment it to not have such large data
        for v in lt_d:              # this for loop is the process of actions to take on the neighboring nodes
            if colors[v] == 'white':    # if the node is white we turn it gray and add 1 to the distance 
                colors[v] = 'gray'
                d[v] = d[u]+1; parents[v] = u;
                Q.append(v);
        colors[u] = 'black'            # once v is appended to the queue we turn the node black
        
        

BFS(lt_d, 'AT5G49450')




```
